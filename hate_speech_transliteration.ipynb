```python
from nltk.corpus import words
import string
import nltk
```

```python
from googletrans import Translator
from ai4bharat.transliteration import XlitEngine
translator = Translator()
L = 'hi'
e = XlitEngine(L)
```

```python
import pandas as pd
```

```python
import pandas as pd
from glob import glob

file_train = "data/mathur dataset/hi_en_cdmx.txt"
import csv
df_train = pd.read_csv(file_train,sep=",",encoding='utf-8',quoting=csv.QUOTE_NONE,usecols=[0,1])

df_train = df_train.dropna()

print(df_train.info())

train_sentences = list(df_train['text'].values)
train_labels = list(df_train['hate'].values)
```

```python
dict_word_lang=[]
total_transliterated_sentences = []
new_labels = []
i=0
for line,label in zip(train_sentences,train_labels):
    i+=1
    print(i)
    #line = line.translate(str.maketrans('', '', string.punctuation))
    line = line.strip()
    line_list = line.split()
    print(line_list)
    new_line_list = []
    for word in line_list:
        #print(word)
        if "@" in word:
            new_line_list.append(word)
        elif "http" in word:
            new_line_list.append(word)
        elif word.lower() in words.words():
            #print(word,"is english word")
            value = translator.translate(word,dest=L).text
            #print(value)
            new_line_list.append(value)
        elif word.isdigit():
            new_line_list.append(word)
        else:
            out = e.translit_word(word, beam_width=10)
            value = out[L][0]
            #print(word,out['ta'][0],"transliterated word")
            new_line_list.append(value)
    print(" ".join(new_line_list))

    if len(line_list) == len(new_line_list):
        print(True)
    else:
        print(line_list)
    total_transliterated_sentences.append(new_line_list)
    new_labels.append(label)
```

```python
new_total_transliterated_sentences = []
for line in total_transliterated_sentences:
    new_total_transliterated_sentences.append(" ".join(line))
```

```python
train_transliterated = pd.DataFrame(zip(new_total_transliterated_sentences,train_labels),columns=['text','label'])
train_transliterated.to_csv("mathur_dataset_hi_en_cdmx_transliterated.csv",index=False)
```
