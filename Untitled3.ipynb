```python
import pandas as pd
df = pd.read_csv("hate_speech_hindi_english_transliterated.tsv",sep="\t")
```

```python
sentences = df['text'].values
```

```python
final_words=[]
for line in sentences:
    line_list = line.split()
    #print(line_list)
    final_words.extend(line_list)
    
words_set = list(set(final_words))
print(words_set)
```

```python
for word in words_set[:500]:
    print(word)
```
